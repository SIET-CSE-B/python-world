```python
def options(n):
    if n==2:
        return 'abc'
    if n==3:
        return 'def'
    if n==4:
        return 'ghi'
    if n==5:
        return 'jkl'
    if n==6:
        return 'mno'
    if n==7:
        return 'pqrs'
    if n==8:
        return 'tuv'
    if n==9:
        return 'wxyz'
def keypad(n):
    if n==0:
        output=[]
        output.append('')
        return output
    small=n//10
    last=n%10
    smallout=keypad(small)
    option=options(last)
    second=[]
    for i in smallout:
        for j in option:
            o=i+j
            second.append(o)
    return second

n = int(input())
ans = keypad(n)
for s in ans:
    print(s)
```

```python
def sub(s,o):
    if len(s)==0:
        print(o)
        return
    sub(s[1:],o)
    sub(s[1:],o+s[0])


#main
sub(input(),"")
```
