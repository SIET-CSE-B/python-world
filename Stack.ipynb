```python
def checkRedundant(string):
    if len(string)==0:
        return False
    b1='('
    b2=')'
    c=0
    v=0
    c1=0
    for i in range(len(string)):
        if string[i] in b1:
            c=c+1
        if string[i] in b2:
            v=v+1
        if string[i] in b1 and string[i+1] in b2:
            return True
        if string[i] in b1 and string[i+1] in b1:
            return True
        if string[i] not in b1 and string[i] not in b2:
            c1=c1+1
        if c==v and c1!=0 and len(string)-1==i:
            return False
            
        elif c==v:
            c1=0
            c=0
            v=0
        else:
            pass
    if c1==0:
        return True
    else:
        return False
    

string = input()
ans = checkRedundant(string)
if ans is True:
    print('true')
else:
    print('false')
```

```python
def stockSpan(price,n):
    s=[]
    o=[]
    s.append(0)
    o.append(1)
    for i in range(1,len(price)):
        while (len(s)!=0 and price[s[-1]]<price[i]):
            s.pop()
        if len(s)==0:
            o.append(i+1)
        else:
            o.append(i-s[-1])
        s.append(i)
    return o
from sys import setrecursionlimit
setrecursionlimit(11000)  
n = int(input())
price = [int(ele) for ele in input().split()]
spans = stockSpan(price,n)
for ele in spans:
    print(ele,end= ' ')
```
